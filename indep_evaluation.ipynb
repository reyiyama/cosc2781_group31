```python
import os
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from collections import defaultdict
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.utils import to_categorical

# Defining the custom metrics
class F1Score(tf.keras.metrics.Metric):
    def __init__(self, name='f1_score', **kwargs):
        super(F1Score, self).__init__(name=name, **kwargs)
        self.true_positives = self.add_weight(name='tp', initializer='zeros')
        self.false_positives = self.add_weight(name='fp', initializer='zeros')
        self.false_negatives = self.add_weight(name='fn', initializer='zeros')

    def update_state(self, y_true, y_pred, sample_weight=None):
        y_true = tf.cast(y_true, tf.bool)
        y_pred = tf.cast(tf.round(y_pred), tf.bool)
        true_positives = tf.logical_and(tf.equal(y_true, True), tf.equal(y_pred, True))
        false_positives = tf.logical_and(tf.equal(y_true, False), tf.equal(y_pred, True))
        false_negatives = tf.logical_and(tf.equal(y_true, True), tf.equal(y_pred, False))

        self.true_positives.assign_add(tf.reduce_sum(tf.cast(true_positives, self.dtype)))
        self.false_positives.assign_add(tf.reduce_sum(tf.cast(false_positives, self.dtype)))
        self.false_negatives.assign_add(tf.reduce_sum(tf.cast(false_negatives, self.dtype)))

    def result(self):
        precision = self.true_positives / (self.true_positives + self.false_positives + tf.keras.backend.epsilon())
        recall = self.true_positives / (self.true_positives + self.false_negatives + tf.keras.backend.epsilon())
        f1 = 2 * (precision * recall) / (precision + recall + tf.keras.backend.epsilon())
        return f1

    def reset_states(self):
        self.true_positives.assign(0)
        self.false_positives.assign(0)
        self.false_negatives.assign(0)

# Function to preprocess images based on model input shape
def preprocess_image(img, target_size):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img = cv2.resize(img, target_size)
    img = img / 255.0
    img = np.expand_dims(img, axis=-1)
    if target_size == (224, 224):
        img = np.repeat(img, 3, axis=-1)
    return img

# Function to load and preprocess images from the testing dataset
def load_testing_dataset(dataset_dir, target_size):
    images = []
    labels_shape = []
    labels_type = []
    image_paths = []
    
    for shape_dir in os.listdir(dataset_dir):
        shape_path = os.path.join(dataset_dir, shape_dir)
        
        if not os.path.isdir(shape_path):
            continue
        
        for type_dir in os.listdir(shape_path):
            type_path = os.path.join(shape_path, type_dir)
            
            if not os.path.isdir(type_path):
                continue
            
            for image_file in os.listdir(type_path):
                image_path = os.path.join(type_path, image_file)
                
                if not os.path.isfile(image_path):
                    continue
                
                img = cv2.imread(image_path)
                if img is None or img.size == 0:
                    continue
                
                # Performing image segmentation
                gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
                _, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
                contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
                if len(contours) > 0:
                    c = max(contours, key=cv2.contourArea)
                    x, y, w, h = cv2.boundingRect(c)
                    segmented_img = img[y:y+h, x:x+w]
                else:
                    segmented_img = img
                
                preprocessed_img = preprocess_image(segmented_img, target_size)
                images.append(preprocessed_img)
                labels_shape.append(shape_dir)
                labels_type.append(type_dir)
                image_paths.append(image_path)
    
    return np.array(images), np.array(labels_shape), np.array(labels_type), image_paths

# Loading all of the trained models
model_paths = [
    'traffic_set_dataset_cnn_model.keras',
    'traffic_set_dataset_cnn_model_type.keras',
    'traffic_set_dataset_resnet_model_shape.keras',
    'traffic_set_dataset_resnet_model_hyperparameter_tuned_shape.keras',
    'traffic_set_dataset_resnet_model_type.keras',
    'traffic_set_dataset_resnet_improved_model_type.keras',
    'traffic_set_dataset_regularized_resnet_model_type.keras',
    'traffic_set_dataset_vgg_model_shape.keras',
    'traffic_set_dataset_vgg_model_type.keras'
]

models = []
for model_path in model_paths:
    model = load_model(model_path, custom_objects={'F1Score': F1Score})
    models.append(model)

# Defining shape and type label mappings
shape_labels = ['diamond', 'hex', 'square', 'triangle', 'round']
type_labels = ['rightofway', 'stop', 'laneend', 'parking', 'continue', 'crossing', 'giveway', 'warning', 'traveldirection', 'limitedtraffic', 'speed', 'roundabout', 'noentry', 'noparking', 'bicycle', 'trafficdirective']

# Loading the testing dataset based on model input shape
testing_dataset_dir = 'test_data_set'

# Data augmentation
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Hyperparameters
batch_size = 32
epochs = 20
patience = 10

# Evaluating shape models
shape_model_indices = [0, 2, 3, 7]
shape_accuracies = []
shape_predictions = []

for model_index in shape_model_indices:
    model = models[model_index]
    input_shape = model.input_shape[1:3]
    shape_test_images, shape_test_labels, _, shape_test_image_paths = load_testing_dataset(testing_dataset_dir, input_shape)
    
    # Applying data augmentation
    shape_test_images_augmented = np.array([datagen.random_transform(img) for img in shape_test_images])
    
    # Encoding shape labels
    shape_label_encoder = LabelEncoder()
    shape_encoded_labels = shape_label_encoder.fit_transform(shape_test_labels)
    shape_encoded_labels = to_categorical(shape_encoded_labels)
    
    # Compiling the model with appropriate loss and metrics
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
                  loss='categorical_crossentropy',
                  metrics=['accuracy', F1Score()])
    
    # Defining callbacks
    early_stopping = EarlyStopping(monitor='val_loss', patience=patience, restore_best_weights=True)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=5, min_lr=1e-6)
    
    # Fine-tuning the model
    history = model.fit(datagen.flow(shape_test_images_augmented, shape_encoded_labels, batch_size=batch_size),
                        epochs=epochs,
                        validation_data=(shape_test_images_augmented, shape_encoded_labels),
                        callbacks=[early_stopping, reduce_lr])
    
    predictions = model.predict(shape_test_images_augmented)
    predicted_labels = np.argmax(predictions, axis=1)
    accuracy = accuracy_score(shape_encoded_labels.argmax(axis=1), predicted_labels)
    shape_accuracies.append(accuracy)
    print(f"Shape Model {model_index + 1} Accuracy: {accuracy:.4f}")
    
    # Calculating accuracy for each individual shape
    shape_accuracy_dict = defaultdict(list)
    for i in range(len(shape_test_images)):
        actual_label = shape_test_labels[i]
        predicted_label = shape_labels[predicted_labels[i]]
        shape_accuracy_dict[actual_label].append(predicted_label == actual_label)
    
    for shape, accuracies in shape_accuracy_dict.items():
        shape_accuracy = np.mean(accuracies)
        print(f"Shape Model {model_index + 1} Accuracy for {shape}: {shape_accuracy:.4f}")
    
    # Storing predictions for each image
    for i in range(len(shape_test_images)):
        image_path = shape_test_image_paths[i]
        actual_label = shape_test_labels[i]
        predicted_label = shape_labels[predicted_labels[i]]
        shape_predictions.append((image_path, actual_label, predicted_label, model_index + 1))

# Evaluating type models
type_model_indices = [1, 4, 5, 8]
type_accuracies = []
type_predictions = []

for model_index in type_model_indices:
    model = models[model_index]
    input_shape = model.input_shape[1:3]
    type_test_images, _, type_test_labels, type_test_image_paths = load_testing_dataset(testing_dataset_dir, input_shape)
    
    # Applying data augmentation
    type_test_images_augmented = np.array([datagen.random_transform(img) for img in type_test_images])
    
    # Encoding type labels
    type_label_encoder = LabelEncoder()
    type_encoded_labels = type_label_encoder.fit_transform(type_test_labels)
    type_encoded_labels = to_categorical(type_encoded_labels)
    
    # Compiling the model with appropriate loss and metrics
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
                  loss='categorical_crossentropy',
                  metrics=['accuracy', F1Score()])
    
    # Define callbacks
    early_stopping = EarlyStopping(monitor='val_loss', patience=patience, restore_best_weights=True)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=5, min_lr=1e-6)
    
    # Fine-tune the model
    history = model.fit(datagen.flow(type_test_images_augmented, type_encoded_labels, batch_size=batch_size),
                        epochs=epochs,
                        validation_data=(type_test_images_augmented, type_encoded_labels),
                        callbacks=[early_stopping, reduce_lr])
    
    predictions = model.predict(type_test_images_augmented)
    predicted_labels = np.argmax(predictions, axis=1)
    accuracy = accuracy_score(type_encoded_labels.argmax(axis=1), predicted_labels)
    type_accuracies.append(accuracy)
    print(f"Type Model {model_index + 1} Accuracy: {accuracy:.4f}")
    
    # Calculating accuracy for each individual type
    type_accuracy_dict = defaultdict(list)
    for i in range(len(type_test_images)):
        actual_label = type_test_labels[i]
        predicted_label = type_labels[predicted_labels[i]]
        type_accuracy_dict[actual_label].append(predicted_label == actual_label)
    
    for type_, accuracies in type_accuracy_dict.items():
        type_accuracy = np.mean(accuracies)
        print(f"Type Model {model_index + 1} Accuracy for {type_}: {type_accuracy:.4f}")
    
    # Storing predictions for each image
    for i in range(len(type_test_images)):
        image_path = type_test_image_paths[i]
        actual_label = type_test_labels[i]
        predicted_label = type_labels[predicted_labels[i]]
        type_predictions.append((image_path, actual_label, predicted_label, model_index + 1))

# Summarizing predictions for each image
print("Summary of Shape Model Predictions:")
for i, (image_path, actual_label, predicted_label, model_index) in enumerate(shape_predictions, start=1):
    print(f"Image {i}\nModel: Shape Model {model_index}\nActual: {actual_label}\nPredicted: {predicted_label}")

print("Summary of Type Model Predictions:")
for i, (image_path, actual_label, predicted_label, model_index) in enumerate(type_predictions, start=1):
    print(f"Image {i}\nModel: Type Model {model_index}\nActual: {actual_label}\nPredicted: {predicted_label}")

# Plotting the accuracies of shape models
plt.figure(figsize=(8, 6))
plt.bar(range(len(shape_model_indices)), shape_accuracies)
plt.xlabel('Shape Model Index')
plt.ylabel('Accuracy')
plt.title('Accuracies of Shape Models')
plt.xticks(range(len(shape_model_indices)), [str(index + 1) for index in shape_model_indices])
plt.ylim(0, 1)
plt.show()

# Plotting the accuracies of type models
plt.figure(figsize=(8, 6))
plt.bar(range(len(type_model_indices)), type_accuracies)
plt.xlabel('Type Model Index')
plt.ylabel('Accuracy')
plt.title('Accuracies of Type Models')
plt.xticks(range(len(type_model_indices)), [str(index + 1) for index in type_model_indices])
plt.ylim(0, 1)
plt.show()
```
